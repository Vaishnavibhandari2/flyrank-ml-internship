## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
why_ml = """
A fixed rule such as 'refresh every page with CTR below 0.4'
is too simple because content performance depends on many factors.

Machine learning can combine multiple signals like CTR,
impressions, traffic trend and engagement to assign
a better refresh priority score.

This helps prioritize the pages that are most likely
to improve after being refreshed.
"""

print(why_ml)
```
